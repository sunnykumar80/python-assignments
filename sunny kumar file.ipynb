```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt


S = [5*np.sin(19*np.pi*t) for t in np.linspace(0,1, 1000)]

U = [7*np.sin(21*np.pi*t) for t in np.linspace(0,3, 1000)]

N = [9*np.sin(14*np.pi*t) for t in np.linspace(0,4, 1000)]

output=[i+J+K for i,J,K in zip(S,U,N)]

plt.plot(output)

Sun=np.fft.fft(output)
plt.plot(output)
print(output)
plt.plot(Sun)
plt.plot(Sun)
plt.show()
```
